```python
import json
import urllib.request

users = [
            'elmiram', 'maryszmary', 'lizaku', 'nevmenandr', 'ancatmara', 'roctbb',
            'akutuzov', 'agricolamz', 'lehkost', 'kylepjohnson', 'mikekestemont',
            'demidovakatya', 'shwars', 'JelteF', 'timgraham', 'arogozhnikov',
            'jasny', 'bcongdon', 'whyisjake', 'gvanrossum'
        ]


def in_func():
    print(str(users[:5]) + '\n' + str(users[5:10]) + '\n' + str(users[10:15]) + '\n' + str(users[15:20]))
    a = input("Введите юзера из списка:")
    if a in users:
        print("Вы выбрали юзера ", a)
    else:
        print("Значения нет в списке")
        a = input("Введите заново:")
    return a


def main():
    url = 'https://api.github.com/users/%s/repos' % in_func()
    response = urllib.request.urlopen(url)
    text = response.read().decode('utf-8')
    data = json.loads(text)
    print('Вот список его репозиториев их описанием:')
    for i in data:
        print(i["name"], ':', i["description"])
        print(i["language"])
    return i


if __name__ == '__main__':
    main()
```

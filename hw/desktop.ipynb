```python
import random

# В работе использован кусок код виселицы с сайта writegamepython.wordpress.com

# Создает рисунки виселицы, а заодно устанавливает количество возможно допустимых ошибок
pics = ['''
    T-------T
            |
            |
            |
            |
            |
            |
     _ _ _ _J''', '''


    T-------T
    |       |
            |
            |
            |
            |
            |
     _ _ _ _J''', '''


    T-------T
    |       |
    Q       |
            |
            |
            |
            |
     _ _ _ _J''', '''


    T-------T
    |       |
    Q       |
    |       |
            |
            |
            |
     _ _ _ _J''', '''


    T-------T
    |       |
    Q       |
    |\      |
            |
            |
            |
     _ _ _ _J''', '''


     T------T
     |      |
     Q      |
    /|\     |
            |
            |
            |
     _ _ _ _J''', '''


     T------T
     |      |
     Q      |
    /|\     |
      \     |
            |
            |
     _ _ _ _J''', '''


     T------T
     |      |
     Q      |
    /|\     |
    / \     |
            |
            |
     _ _ _ _J''']


def choose_library():
    # Функция позволяет юзеру выбрать библиотеку слов
    i = input("Выберите библиотеку слов: " + "\n" + "1.Киножанры" +
              "\n" + "2.Киноактеры" + "\n" + "3.Кинорежиссеры"+ "\n" + "Ваш выбор:")
    if i == '1':
        with open('genres.txt', encoding='utf-8') as f:
            lib = f.read()
            lib = lib.split(' ')
            print('Вы выбрали библиотеку "киножанры"')
    elif i == '2':
        with open('actors.txt', encoding='utf-8') as f:
            lib = f.read()
            lib = lib.split(' ')
            print('Вы выбрали библиотеку "киноактеры"')
    elif i == '3':
        with open('directors.txt', encoding='utf-8') as f:
            lib = f.read()
            lib = lib.split(' ')
            print('Вы выбрали библиотеку "кинорежиссеры"')
    elif i != '1' and i != '2' and i != '3':
        print('Повторите ввод.')
        with open('music.txt', encoding='utf-8') as f:
            lib = f.read()
            lib = lib.split(' ')
            print('Системный выбор. Секретная библиотека')

    indexs = random.randint(0, len(lib))
    return lib[indexs]


def wordworks(pics, correct, incorrect, quiz):
    # Здесь функция выводит рисунок виселицы, соответствующий количеству ошибок
    print(pics[len(incorrect)])
    print()
    if len(incorrect) in range(1):
        print('У вас есть 8 попыток, чтобы отгадать слово из ' + str(len(quiz)) + ' букв.')
    # А здесь она показывает ошибки юзера
    print('Этих букв в слове нет: ', end=' ')
    for letter in incorrect:
        print(letter, end=' ')
    print()
    # Заменяет буквы слова на нижние подчеркивания
    blanks = '_'*len(quiz)
    # Заменяет подчеркивания на правильно угаданные буквы
    for i in range(len(quiz)):
        if quiz[i] in correct:
            blanks = blanks[:i] + quiz[i] + blanks[i+1:]
    for letter in blanks:
        print(letter, end=' ')
    print()


def check(usersletters):
    # Проверяет корректность введенных букв
    while True:
        guess = input(print('Введите букву:'))
        guess = guess.lower()
        if len(guess) != 1:
            print('Неверное значение.')
        elif guess in usersletters:
            print('Буква уже введена.')
        elif guess not in 'йцукенгшщзхъфывапролджёэячсмитьбю':
            print('Не кириллический символ.')
        else:
            return guess


def onemore():
    # Спрашивает, хочет ли юзер сыграть еще раз
    answer = input(print('Попробовать еще раз?'))
    while True:
        answer = answer.startswith('д')
        return answer
    

# Создаем списки

incorrect = ''
correct = ''
quiz = choose_library()
finish = False

while True:
    wordworks(pics, correct, incorrect, quiz)
    # Проверяем вхождение всех введенных букв в список
    guess = check(correct + incorrect)
    if guess in quiz:
        correct = correct + guess
        allin = len(correct) == len(quiz)
        if allin:
            print('Молодец! Отгадано слово " ' + quiz + '"')
            finish = True
    else:
        incorrect = incorrect + guess
        # Учим игру склонять "попытки"
        left = len(pics) - len(incorrect)
        if len(incorrect) in range(3):
            print('У вас осталось ' + str(left) + ' попыток')
        elif len(incorrect) in range(3, 6):
            print('У вас осталось ' + str(left) + ' попытки')
        elif len(incorrect) in range(6, 7):
            print('У вас осталась 1 попытка')

        if len(incorrect) == len(pics) - 1:
            wordworks(pics, correct, incorrect, quiz)
            print('Игра окончена. Попыток больше не осталось.' + '\n' + 'Загаданное слово:' + quiz)
            finish = True
        if finish:
            if onemore():
                print('Хорошо, запускаю еще раз.')
                incorrect = ''
                correct = ''
                quiz = choose_library()
                finish = False
            else:
                break
```
